```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
#create a figure
f = plt.figure(figsize=(5,10))

#get the turbo color map
cmap_turbo = cm.get_cmap('turbo')

#add a single subplot to the figure
#with empty x anf y ticklabels
#and no axes shown
ax = f.add_subplot(111,xticks=[], yticks=[])
ax.set_axis_off()

#set the limits to x and y axes
ax.set_xlim([0, 2*np.pi])
ax.set_ylim([0, 15.])

#create an x arrange [0, 2pi]
x = np.linspace(0,2*np.pi,100)

#set the number of mountain ranges
nmr = 10
#set the number of peaks in each range
npeak = 10

#loop over the mountain ranges
for i in range(nmr):
    #set the color of this mountain range
    #by providing a number [0,1]
    mr_color = cmap_turbo(float(i)/float(npeak))

    #create a randomized set of frequencies
    freq = 3.0*np.random.rand(npeak) + 1.0

    phi = 2.0*np.pi*np.random.rand(npeak) + 1.0

    amp = 1.*np.random.rand(npeak) + 0.1

    #set a y offset, so each range is
    #offset along the y direction
    yoff = 1.*float(i)

    #create the mountain range
    y = np.zeros_like(x)

    #for every peak
    for j in range(npeak):

        #add this peak as a sine curve with
        #a random amplitude, frequency, and phase
        y += amp[j]*np.sin(freq[j]*x + phi[j])

    #take the absolute balue of the sin curve
    #and add the offset
    y = np.fabs(y) + yoff

    #plot random mountain range
    plt.fill_between(x,y,yoff,zorder=nmr-i, color=mr_color,cmap='turbo')

#save the figure
plt.savefig('mountain_range.png',bbox_inches='tight',pad_inches=0,dpi=600)



```
