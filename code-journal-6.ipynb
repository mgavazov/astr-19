```python
def function(x):
    sinx = np.sin(x) 
    return sinx
```

```python
def function(x):
    cosx = np.cos(x)
    return cosx
```

```python
x = np.linspace(0.0, 2*np.pi, num=1000)
sinx = np.sin(x)
cosx = np.cos(x)


print("  x   | sin(x)| cos(x) ")

for i in range(1000):
    print(f"{x[i]:.3f} | {sinx[i]:.3f} | {cosx[i]:.3f}")
          
                    
```
